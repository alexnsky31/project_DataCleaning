# Преобразование данных

```python
# создаем новый признак для образования соискателей
def get_education(arg):
    arg = ' '.join(arg.split(' ')[:3])
    if 'Высшее' in arg:
        return 'высшее'
    elif 'Неоконченное высшее' in arg:
        return 'неоконченное высшее'
    elif 'Среднее специальное' in arg:
        return 'среднее специальное'
    elif 'Среднее образование' in arg:
        return 'среднее'
# добовляем его в нашу базу
hh['Образование'] = hh['Образование и ВУЗ'].apply(get_education)
# убераем изначальный 
hh = hh.drop('Образование и ВУЗ', axis=1)
# число соискателей с уровнем образование среднее( школьное оьразование)
print('число соискателей с уровнем образование среднее:', hh['Образование'].value_counts()['среднее'])
```

```python
# определяем пол соискателя
def gender(arg):
    arg = ','.join(arg.split(',')[:1])
    if 'Мужчина' in arg:
        return 'М'
    elif 'Женщина' in arg:
        return 'Ж'
#добовляем новый признак     
hh['Пол'] = hh['Пол, возраст'].apply(gender)
# определяем возраст соискателя 
def get_age(arg):
    arg_splitted = arg.split(' ')
    year_words=['год', 'года', 'лет']
    for index, item in enumerate (arg_splitted):
        if item in year_words:
            return int(arg_splitted[index-1])
#добовляем новый признак   
hh['Возраст'] = hh['Пол, возраст'].apply(get_age)
#убераем изначальный 
hh = hh.drop('Пол, возраст', axis=1)

#процент женщин среди соискателей  
print('процент женщин среди соискателей:', round(hh['Пол'].value_counts(normalize=True)['Ж'] * 100, 2))
#средний возраст соискателя 
print('средний возраст соискателя:', round(hh['Возраст'].mean(), 2))
```

```python
#опыт работы . переводим все в месяци 
def experience(arg):
    if arg is np.nan or arg == 'Не указано':
        return None
    year_words=['год', 'года', 'лет']
    month_words=['месяц', 'месяца', 'месяцев']
    arg_splitted = arg.split(' ')[:7]
    years = 0
    months = 0
    for index, item in enumerate (arg_splitted):
        if item in year_words:
            years = int(arg_splitted[index-1])
        if item in month_words:
            months = int(arg_splitted[index-1])
    return int(years*12 + months)

hh['Опыт работы в месяцах'] = hh['Опыт работы'].apply(experience)
#убераем не нужный признак 
hh = hh.drop('Опыт работы', axis=1)
#среднее значение опыта работы 
print('среднее значение опыта работы: ', round(hh['Опыт работы в месяцах'].median()))
```

```python
# определяем готовность к переезду
def relocation(arg):
    if ('не готов к переезду' in arg) or ('не готова к переезду' in arg) :
        return False
    elif 'чоху' in arg:
        return True
    else:
        return True
 # определяем готовность к командировкам   
def business_trip_ready(arg):
    if ('командировка' in arg):
        if ('не готов к командировкам' in arg) or('не готова к командировкам' in arg):
            return False
        else:
            return True
    else:
        return False
# определяем определяем город пребывания соискателя         
def city(s):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань',
                      'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 
                      'Воронеж', 'Волгоград']
    s = s.lower()
    if 'москва' in s:
        return 'Москва'
    elif 'санкт-петербург' in s:
        return 'Санкт-Петербург'
    elif any(city in s for city in million_cities):
        return 'город-миллионник'
    else:
        return 'другие'
#добовляем новые признаки в df вместо исходного ('Город, переезд, командировки)
hh['Город'] = hh['Город, переезд, командировки'].apply(city)
hh['Готовность к переезду'] = hh['Город, переезд, командировки'].apply(relocation)
hh['Готовность к командировкам'] = hh['Город, переезд, командировки'].apply(lambda x: business_trip_ready(x) if type(x) == str else False)
hh = hh.drop('Город, переезд, командировки', axis=1)

#процент соискателей из СПБ и процент готовый (к переезду и командировка)
sp_count = (hh["Город"] == "Санкт-Петербург").sum()
total_count = len(hh)
sp_percent = sp_count / total_count * 100
print(f"Процент соискателей, живущих в Санкт-Петербурге: {sp_percent:.2f}%")
ready_to_trip = round(hh[hh['Готовность к переезду'] & hh['Готовность к командировкам']].shape[0] / hh.shape[0] *100)
print('процент соискателей готовый к переезду и командировка:', round(hh[hh['Готовность к переезду'] & hh['Готовность к командировкам']].shape[0] / hh.shape[0] *100))
```

```python
#преобразовываем признак занятости к полее уникальным признакам
work_mask1 = ['полная занятость', 'частичная занятость',
              'проектная работа', 'волонтерство', 'стажировка']
work_mask2 = ['полный день', 'сменный график', 
         'гибкий график', 'удаленная работа',
         'вахтовый метод']

for work_mask1, work_mask2 in zip(work_mask1, work_mask2):
    hh[work_mask1] = hh['Занятость'].apply(lambda x: work_mask1 in x)
    hh[work_mask2] = hh['График'].apply(lambda x: work_mask2 in x)
#Меняем исходный признак на новые
hh = hh.drop('Занятость', axis=1)
hh = hh.drop('График', axis=1)

display(hh.head())
```

```python
#DF с курсом валют стандартным методом  (Для этого убираем #перед в начале строк )
    #   exchanch = pd.read_csv('E:\SF\project_01\ExchangeRates.csv')
    #   display(exchanch.tail())

# вариант с прямым открытием df из google disk( при условии если фай открым "все у кого есть ссылка")
url='https://drive.google.com/file/d/1KgYQvbKQ9g7347ZZb3WNMS7L1-8PP6gb/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
exchange = pd.read_csv(dwn_url)

# разделяем сумму и валюту
def get_salary_num(arg):
    salary = float(arg.split(' ')[0])
    return salary
#меняем указаную валюту в hh на ISO формат
def get_salary_сurrency(arg):
    currency_dict = {
        'USD': 'USD', 'KZT': 'KZT',
        'грн': 'UAH', 'белруб': 'BYN',
        'EUR': 'EUR', 'KGS': 'KGS',
        'сум': 'UZS', 'AZN': 'AZN'
    }
    curr = arg.split(' ')[1].replace('.', '')
    if curr == 'руб':
        return 'RUB'
    else:
        return currency_dict[curr]
#проводим признаки к формату datetime и пременяем фунции созданые ранее    
exchange['date'] = pd.to_datetime(exchange['date']).dt.date
hh['Обновление резюме'] = pd.to_datetime(hh['Обновление резюме']).dt.date
hh['ЗП (tmp)'] = hh['ЗП'].apply(get_salary_num)
hh['Курс (tmp)'] = hh['ЗП'].apply(get_salary_сurrency)
#объеденяем dataframe's и убераем ненужные признаки 
merged = hh.merge(
    exchange, 
    left_on=['Курс (tmp)', 'Обновление резюме'],
    right_on=['currency', 'date',], 
    how='left'
)
merged['close'] = merged['close'].fillna(1)
merged['proportion'] = merged['proportion'].fillna(1)
hh['ЗП (руб)'] = merged['close'] * merged['ЗП (tmp)'] / merged['proportion']
hh = hh.drop(['ЗП', 'ЗП (tmp)', 'Курс (tmp)'], axis=1)
#смотрим медийное значение 
print('медийное значение ЗП:', round(hh['ЗП (руб)'].median()/1000))

display(hh.head())
```
